## LIBRARIES

```python
import holoviews as hv
from holoviews import dim
hv.extension('bokeh')
from bokeh.plotting import show
from bokeh.io import export_svgs
import pyperclip
import pandas as pd
from collections import Counter
import itertools
import plotly.graph_objects as go
```

## IMPORTING DATA FROM EXCEL CLIPBOARD

```python
df = pd.read_clipboard()
df = df.fillna('N/A')
```

## Re-Order

```python
df1 = df.loc[df['FactorType'] == 'Tracking'].copy()
df2 = df.loc[df['FactorType'] == 'Path'].copy()

df3 = pd.DataFrame(df1['Attribute']).merge(df2, on = 'Attribute', how = 'left').dropna()
df  = pd.concat([df1, df3]).reset_index(drop = True)
check = df[['FactorType', 'Attribute']].groupby('Attribute').count()
not_attr = check.loc[check['FactorType'] < 2].index.to_list()
df = df.loc[~df['Attribute'].isin(not_attr)].reset_index(drop = True)
df
```

## GET LABELS AND COLORS

```python
tracker_attrs = df.loc[df['FactorType'] == "Tracking"]
attrs_path = df.loc[df['FactorType'] == "Path"]

tracker_colors = dict(zip(tracker_attrs["FactorName"], tracker_attrs["HexColor"]))
attr_colors = dict(zip(tracker_attrs["Attribute"], tracker_attrs["HexColorAtt"])) # Colores cajas de atributos 
# attr_colors = dict(zip(tracker_attrs["Attribute"], itertools.repeat("#808080"))) # Cajas grises atributos
path_colors = dict(zip(attrs_path["FactorName"], attrs_path["HexColor"]))
```

```python
# Get unique labels in order of first appearance
tracker_labels = tracker_attrs["FactorName"].drop_duplicates().tolist()
attr_labels = tracker_attrs["Attribute"].drop_duplicates().tolist()

path_labels = attrs_path["FactorName"].drop_duplicates().tolist()

all_labels = tracker_labels + attr_labels + path_labels
```

```python
colormap = []
colormap = list(tracker_colors.values()) +  list(attr_colors.values()) + list(path_colors.values())
print(colormap)
```

## GET SOURCE AND TARGET

```python
mapping_tracker = {item:i for i, item in enumerate(tracker_labels)}
sources = []
targets = []
for item in tracker_attrs["FactorName"]:
    sources.append(mapping_tracker[item])

max_sources = max(sources)

mapping_attrs = {item:i for i, item in enumerate(attr_labels)}

for item in tracker_attrs["Attribute"]:
    sources.append(mapping_attrs[item] + (max_sources+1))
    targets.append(mapping_attrs[item] + (max_sources+1))

max_targets = max(targets)

mapping_path = {item:i for i, item in enumerate(path_labels)}
for item in attrs_path["FactorName"]:
    targets.append(mapping_path[item] + (max_targets+1))


source_target = list(zip(sources, targets))
print(source_target)
```

```python
print(f"{len(sources)}, {len(targets)}")
```

```python
source_target_weight = [(t[0], t[1], 1) for t in source_target]
print(source_target_weight)
```

## PLOT SANKEY

```python
hv.extension("bokeh", "matplotlib")

nodes = all_labels
nodes = hv.Dataset(enumerate(nodes), 'index', 'label')
edges = source_target_weight

value_dim = hv.Dimension('origin')
sankey = hv.Sankey((edges, nodes), ['From', 'To'], vdims=value_dim).opts(
    fontsize={'title': 16, 'labels': 40}, 
    labels='label', 
    label_position='outer', 
    width=1600, 
    height=800, 
    cmap=colormap,
    # edge_color=dim('To').str(), # Color de acuerdo al target 
    edge_color='#E9E9E9', # Color Fijo
    edge_alpha=0.5, # Transparencia de la linea conectora 
    node_color=dim('From').str(), 
    node_width = 15, 
    node_alpha = 10,
    show_values = False,
    node_padding = 10,
    label_text_font_size = '11pt')

show(hv.render(sankey))
```

## PLOT SANKEY USING PLOTLY (PRESERVES NODE ORDER)

```python
# Prepare node colors for Plotly
node_colors_list = []
for label in all_labels:
    if label in tracker_colors:
        node_colors_list.append(tracker_colors[label])
    elif label in attr_colors:
        node_colors_list.append(attr_colors[label])
    elif label in path_colors:
        node_colors_list.append(path_colors[label])
    else:
        node_colors_list.append("#808080")  # fallback color

# Extract source, target, and value from source_target_weight
sources = [t[0] for t in source_target_weight]
targets = [t[1] for t in source_target_weight]
values = [t[2] for t in source_target_weight]

# Create Plotly Sankey
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=all_labels,
        color=node_colors_list
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color="rgba(200, 200, 200, 0.4)"
    )
)])

fig.update_layout(
    title_text="Sankey Diagram (Plotly) - Node Order Preserved",
    font=dict(size=12),
    height=800,
    width=1600
)

fig.show()
```
